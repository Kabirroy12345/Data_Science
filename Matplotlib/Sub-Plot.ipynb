Subplots allow you to show multiple plots in a single figure, side-by-side or in a grid layout. This is helpful when comparing different datasets or aspects of the same data.

```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
```

```python
plt.subplot(1,2,1) # row column kaunsa plot
plt.plot(x,y1)
plt.title("Double of x")
plt.xlabel("Number")
plt.ylabel("Double of number")

plt.subplot(1,2,2) # row column kaunsa plot
plt.plot(x,y2)
plt.title("Square of x")
plt.xlabel("Number")
plt.ylabel("Square of number")
plt.tight_layout()
```

```python
# More variations of x
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]

plt.figure(figsize=(6, 6))  # Optional: make it bigger

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('10 - x')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y1)
axs[0].set_title('x * 2')

axs[1].plot(x, y2)
axs[1].set_title('x squared')
 
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # So title doesn't overlap
fig.savefig('my_plots.png')    # Save as image

plt.show()
```

Figure = the whole sheet of paper
Subplot (Axes) = individual graphs drawn on that sheet
Step 1: Create the figure
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

This is the most confusing line, so let's break it.

plt.subplots(rows, columns)
plt.subplots(1,2)

means

+-------------------------------+
|           Figure              |
|                               |
| +----------+ +----------+     |
| | Graph 1  | | Graph 2  |     |
| +----------+ +----------+     |
+-------------------------------+
1 row
2 columns

So there are 2 graphs side by side.

fig

This stores the whole figure.

fig
┌──────────────────────────┐
│ Entire Window            │

axs

This stores all the subplots.

Since there are two graphs,

axs

is actually

So

axs[0] → Left graph

axs[1] → Right graph
figsize=(10,4)

This controls the size.

Width = 10 inches

Height = 4 inches

A larger width makes the plots wider.

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
```
