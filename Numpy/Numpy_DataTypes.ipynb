```python
import numpy as np
#NumPy arrays are homogeneous, meaning all elements must be of the same type
```

```python
arr=np.array([23,4,3,23])
```

```python
arr.dtype
```

```python
arr.astype("float64") #converting into float from int
```

```python
arr=np.array([23,4,3,23],dtype="float32") #another way
```

```python
arr
```

```python
# The choice of data type affects: - Memory Usage: Smaller data types use less
# memory. - Performance: Operations on smaller data types are faster due to less
# data being processed. - Precision: Choosing the appropriate data type ensures that
# you don’t lose precision (e.g., using float32 instead of float64 if you don’t
# need that extra precision).
```

```python
arr.nbytes # 4 elements the array has be it int or float each is 4 byte
```

```python
arr_int64 = np.array([1, 2, 3], dtype=np.int64)
arr_int32 = np.array([1, 2, 3], dtype=np.int32)
print(arr_int64.nbytes) # Output: 24 bytes (3 elements * 8 bytes each)
print(arr_int32.nbytes) # Output: 12 bytes (3 elements * 4 bytes each)
```

```python
#String Data Type(Usually Not Preffered)
arr = np.array(['apple', 'banana', 'cherry'], dtype='U10') # Unicode string array
print(arr)
#less efficient so dont use uses dtype U for strings
```

```python
arr = np.array([1 + 2j, 3 + 4j, 5 + 6j], dtype='complex128')
print(arr)
#Has 1 real part and 1 complext part and dtype='complex128'
```

```python
arr = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype=object)
print(arr)
#Object data type still not good choice 
```

```python
# Choosing the correct data type is essential for: - Optimizing memory: Using the
# smallest data type that fits your data. - Improving performance: Smaller types
# generally lead to faster operations. - Ensuring precision: Avoid truncating or losing
# important decimal places or values
```
